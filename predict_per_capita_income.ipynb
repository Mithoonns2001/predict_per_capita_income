```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
!wget https://raw.githubusercontent.com/codebasics/py/master/ML/1_linear_reg/Exercise/canada_per_capita_income.csv
```

```python
df= pd.read_csv('/content/canada_per_capita_income.csv')
```

```python
df
```

```python
plt.xlabel("Year")
plt.ylabel("per capita income", color='green')
plt.scatter(df.year, df['per capita income (US$)'], color='red', marker='*')
```

```python
new_df= df.drop('per capita income (US$)', axis='columns')
```

```python
new_df
```

```python
capita=df['per capita income (US$)']
capita
```

```python
model=linear_model.LinearRegression()
model.fit(new_df,capita)
```

```python
# predict

model.predict([[2020]])
```

```python
model.coef_
```
